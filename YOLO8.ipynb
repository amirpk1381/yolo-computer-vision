```python
import cv2
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n-seg.pt')
```

```python
cap = cv2.VideoCapture('df4393fed1299f501832e4eb9f41be2023012684-480p.mp4')
```

```python
while cap.isOpened():
    succes , frame = cap.read()

    if succes:
        results = model(frame)

        annoted_frame = results[0].plot()

        cv2.imshow('yolov8 ' , annoted_frame)

        if cv2.waitKey(1)  & 0xff == ord('q'):
            break

    else:
        break
           

```

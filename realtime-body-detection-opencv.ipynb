```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture('df4393fed1299f501832e4eb9f41be2023012684-480p.mp4')
```

```python
body_classifier = cv2.CascadeClassifier('haarcascade_fullbody.xml')
```

```python
ret , frame = cap.read()
```

```python
frame[0]
```

```python
if ret:
    gray=cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    body=body_classifier.detectMultiScale(gray  , 1.2 , 3)
    for x,y,w,h in body:
        cv2.rectangle(frame , (x,y) , (x+w,y+h) , (127,255,0) ,2)

cap.release()
imshow('cc' , frame)        
```
